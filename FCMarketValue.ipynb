```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
data = pd.read_csv('club.csv')
```

```python
data.head()
```

```python
data.shape
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
df1 = data['Competition Name'].value_counts().reset_index()

plt.figure(figsize=(15,6))
sns.barplot(x="index", y="Competition Name", data=df1)
plt.title("No. of teams in every League")
plt.xlabel("Competition Name")
plt.ylabel("Number of Competition")
plt.xticks(rotation=80)
plt.show()
```

```python
df2 = data.groupby(["Club Name","Competition Name"]).mean().reset_index().drop("Unnamed: 0", axis=1)
```

```python
df2.head()
```

```python
plt.figure(figsize=(44,14))
sns.barplot(y="Market Value Of Club In Millions(£)", x="Club Name", data=df2.sort_index())
plt.title("RElation b/w Club and Mean Market Value of Club in Millions(£)")
plt.xticks(rotation=80)
plt.show()
```

```python
plt.figure(figsize=(44,14))
sns.barplot(y="Average Age Of Players", x="Club Name", data=df2.sort_index())
plt.title("Relation b/w Club and Average Age of the Players")
plt.xticks(rotation=80)
plt.show()
```

```python
plt.figure(figsize=(22,7))
sns.barplot(y = "Average Market Value Of Players In Millions(£)", hue="Squad Size", x="Competition Name", data=df2.sort_index())
plt.title("Relation b/w Average Market Value of  Players in Million(£)) and Squad Size")
plt.xticks(rotation=80)
plt.show()
```

```python
plt.figure(figsize=(44,14))
sns.barplot(y = "Squad Size", x="Club Name", data = df2.sort_index())
plt.title("Relation b/w Club Name and Squad Size")
plt.xticks(rotation=80)
plt.show()
```

Conclusions



*   **Premier League** is most held competition.
*   **Manchester City** and **Paris Saint-Germain** have the most market value in Million(£).
*   Average age of players is **25-30 yrs.**
*   **Premier League** has the most market value of players.
*   Average Squad size is **25-30** but **PSG** has max squad size which is **35**.
